```python
import os
from rdflib import Graph,Literal, RDF, URIRef
from rdflib.namespace import FOAF, XSD
from rdflib.namespace import  RDF, RDFS
```

```python
import xml.etree.ElementTree as ET
```

```python
xml_path = "Family.xml"
onto_path = "CS22M086_Family_Ontology.ttl"
```

```python
tree_ele = ET.parse(xml_path)
root_ele = tree_ele.getroot()
```

```python
graph=Graph()
graph.parse(onto_path)
```

```python
myNamespace="http://www.semanticweb.org/sudhanshu/ontologies/2023/3/FamilyOntology"
rdftype = URIRef("http://www.w3.org/1999/02/22-rdf-syntax-ns#type")
```

```python
def isAlreadyDefined(subs):
    for s in graph.subjects():
        if(subs in str(s)):
            return True
    return False
```

```python
isFatherOf = []
isMotherOf = []
isSonOf = []
isDaughterOf = []
isHusbandOf = []
isWifeOf = []
isBrotherOf = []
isSisterOf = []
Male = []
Female = []
```

```python
male = URIRef(str(myNamespace)+"/Male")
female = URIRef(str(myNamespace)+"/Female")
isfatherof = URIRef(str(myNamespace)+"/isFatherOf")
ismotherof = URIRef(str(myNamespace)+"/isMotherOf")
issonof = URIRef(str(myNamespace)+"/isSonOf")
isdaughterof = URIRef(str(myNamespace)+"/isDaughterOf")
ishusbandof = URIRef(str(myNamespace)+"/isHusbandOf")
iswifeof = URIRef(str(myNamespace)+"/isWifeOf")
issisterof = URIRef(str(myNamespace)+"/isSisterOf")
isbrotherof = URIRef(str(myNamespace)+"/isBrotherOf")
```

```python

for family in root_ele:
    for person in family:
        son = []
        daughter = []
        name = person[0].text
        name = name.replace(" ","_") 
        name = URIRef(str(myNamespace) + "/" +str(name))
        gender = person[1].text
        spouse = person[4].text
        spouse = spouse.replace(" ","_")
        spouse = URIRef(str(myNamespace) + "/" +str(spouse))
        for x in person:
            if(x.tag == "Son"):
                son.append(x.text)
            if(x.tag == "Daughter"):
                daughter.append(x.text)
        if(gender == "Male"):
            Male.append((name,RDF.type, male))
            for x in son:
                x = x.replace(" ","_")
                son_name = URIRef(str(myNamespace) + "/" +str(x))
                isFatherOf.append((name,isfatherof,son_name))
                isSonOf.append((son_name,issonof,name))
            for y in daughter:
                y = y.replace(" ","_")
                daughter_name = URIRef(str(myNamespace) + "/" +str(y))
                isFatherOf.append((name,isfatherof,daughter_name))
                isDaughterOf.append((daughter_name,isdaughterof,name))
            isHusbandOf.append((name,ishusbandof,spouse))
            
        if(gender == "Female"):
            Female.append((name,RDF.type, female))
            for x in son:
                x = x.replace(" ","_")
                son_name = URIRef(str(myNamespace) + "/" +str(x))
                isMotherOf.append((name,ismotherof,son_name))
                isSonOf.append((son_name,issonof,name))
            for y in daughter:
                y = y.replace(" ","_")
                daughter_name = URIRef(str(myNamespace) + "/" +str(y))
                isMotherOf.append((name,ismotherof,daughter_name))
                isDaughterOf.append((daughter_name,isdaughterof,name))
            isWifeOf.append((name,iswifeof,spouse))
        for x in son:
            x = x.replace(" ","_")
            for xx in son:
                xx = xx.replace(" ","_")
                if(x!=xx):
                    x_name = URIRef(str(myNamespace) + "/" +str(x))
                    xx_name = URIRef(str(myNamespace) + "/" +str(xx))
                    isBrotherOf.append((x_name,isbrotherof,xx_name))
            for xx in daughter:
                xx = xx.replace(" ","_")
                if(x!=xx):
                    x_name = URIRef(str(myNamespace) + "/" +str(x))
                    xx_name = URIRef(str(myNamespace) + "/" +str(xx))
                    isBrotherOf.append((x_name,isbrotherof,xx_name))
        for x in daughter:
            x = x.replace(" ","_")
            for xx in son:
                xx = xx.replace(" ","_")
                if(x!=xx):
                    x_name = URIRef(str(myNamespace) + "/" +str(x))
                    xx_name = URIRef(str(myNamespace) + "/" +str(xx))
                    isSisterOf.append((x_name,issisterof,xx_name))
            for xx in daughter:
                xx = xx.replace(" ","_")
                if(x!=xx):
                    x_name = URIRef(str(myNamespace) + "/" +str(x))
                    xx_name = URIRef(str(myNamespace) + "/" +str(xx))
                    isSisterOf.append((x_name,issisterof,xx_name))
```
